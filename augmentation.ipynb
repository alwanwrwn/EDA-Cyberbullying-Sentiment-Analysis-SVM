```python
def noRs(n, y):
    start = time.time()

    command = "python code/augment.py --input=data/data_training_90.txt --output=data_training_90_" + str(n) +"_" + str(y[0])+ "_NO_RS.txt --num_aug="+ str(n) +" --alpha_sr=" + str(y[1])+ " --alpha_rd=" + str(y[1])+ " --alpha_ri=" + str(y[1])+ " --alpha_rs=0"
    result = subprocess.run(command, shell=True, capture_output=True, text=True)

    elapsed = (time.time() - start)/60

    # Print the output
    print(result.stdout)
    print(elapsed)
    print("\n")
```

```python
def run_threads():
    alpha = [['f', 0.5]]
    arguments = [1]  # List of arguments for the function
    threads = []
        
    for arg in arguments:
        for alp in alpha:
            thread = threading.Thread(target=noRs, args=(arg,alp))
            threads.append(thread)
            thread.start()

    # Wait for all threads to complete
    for thread in threads:
        thread.join()

if __name__ == "__main__":
    run_threads()
```
